```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
X = df.drop(columns=['per capita income (US$)'])
y = df['per capita income (US$)']
```

```python
X
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
print('Model Coefficient: ',model.coef_)
print('Model Intercept: ',model.intercept_)
```

```python
income_pred = model.predict(x_test)
income_pred
```

```python
model.predict([[2000]])
```

```python
print('MSE: ',mean_squared_error(y_test, income_pred))
print('R2 Score: ',r2_score(y_test, income_pred))
print('RMSE: ',np.sqrt(mean_squared_error(y_test, income_pred)))
```

```python
# Combine x_train and x_test to cover the full range of data
x_full = np.append(x_train, x_test)
# Sort x_full to ensure the line is smooth and in order
x_full_sorted = np.sort(x_full)
# Predict values for the full range of x_full_sorted
y_full_pred = model.predict(x_full_sorted.reshape(-1, 1))


plt.scatter(x_train, y_train, label='Training Data')
plt.plot(x_full_sorted, y_full_pred, color='r', label='Regression Line')
plt.title('Income per Capita vs Year')
plt.xlabel('Year')
plt.ylabel('Income per Capita')
plt.legend()
plt.show()
```

```python
plt.scatter(x_test, y_test, label='Test Data')
plt.plot(x_test, income_pred, color = 'r', label='Regression Line')
plt.title('Income per Capita vs Year')
plt.xlabel('Year')
plt.ylabel('Income per Capita')
plt.legend()
plt.show()
```
